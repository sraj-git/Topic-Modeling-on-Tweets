```python
#importing libraries
import pandas as pd
import gensim
from gensim import corpora
from gensim.utils import simple_preprocess
```

```python
#loading dataset
processeddata = open("tweets[processed_hindi].csv", "r")
```

```python
processeddata = processeddata.read().decode("utf-8-sig")
```

```python
#appending all hindi tweets in a list
hindi_tweets=[]
for line in processeddata.split('\n'):
    hindi_tweets.append(line)
```

```python
for l in hindi_tweets:
    print l
```

```python
def tokenize(data):
    result = [line.split() for line in data]
    return result
```

```python
processedcorpus=tokenize(hindi_tweets)
```

```python
#Creation of dictionary of processedcorpus, each term is assigned to index
dictionary = corpora.Dictionary(processedcorpus)
```

```python
for i, j in dictionary.items():
    print i, j
```

```python
#filtering out extreme values
dictionary.filter_extremes(no_below=30, no_above=0.5, keep_n=1000)
```

```python
print dictionary.token2id
```

```python
for i, j in dictionary.items():
    print i, j
```

```python
#term document frequency
#conversion of processedccorpus into document term matrix using dictionary
bow_corpus = [dictionary.doc2bow(line) for line in processedcorpus]
```

```python
bow_corpus
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(bow_corpus, num_topics=10, id2word = dictionary, passes=50)
```

```python
for i, words in ldamodel.print_topics():
    print('topic:{} \nlist_of_words: {}'.format( i, words.encode('utf-8')))
    print('\n')
```

```python
#evaluation on the basis of perplexity
print('Perplexity: ', ldamodel.log_perplexity(bow_corpus))
```
