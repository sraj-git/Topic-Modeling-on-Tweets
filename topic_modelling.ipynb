```python
#apply topic modeling
import pandas as pd
import gensim
from gensim.utils import simple_preprocess
```

```python
processeddata = pd.read_csv('tweets[processed_english].csv')
```

```python
#tokenizing the processed corpus
def tokenize(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
            result.append(token)
    return result
```

```python
#map is used to tokenize
processed_corpus = processeddata['text'].map(tokenize)
```

```python
print processed_corpus[0:10]
```

```python
#bag of word on the dataset
dictionary = gensim.corpora.Dictionary(processed_corpus)
```

```python
print dictionary
```

```python
dictionary.filter_extremes(no_below=40, keep_n=600)
```

```python
print dictionary
```

```python
print dictionary.token2id
```

```python
#term document frequency
bow_corpus = [dictionary.doc2bow(text) for text in processed_corpus]
```

```python
bow_corpus
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=bow_corpus,
                                           id2word=dictionary,
                                           num_topics=10, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=200,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
for i, words in lda_model.print_topics():
    print('topic:{} \nlist_of_words: {}'.format( i, words))
    print('\n')
```

```python
test_document = "mobile sim number linked to adhaar"
test_bow_vector = dictionary.doc2bow(tokenize(test_document))
test_bow_vector
```
